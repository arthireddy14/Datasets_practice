Task 2 — Clean the Dataset
Identify and count missing values in each column.
Drop column(s) that are irrelevant for price analysis (example: society or unnamed columns).
Handle missing values in numeric columns such as bath, balcony, etc. (use either dropna() or fillna() depending on your reasoning).

```python

df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=False)
```

```python
df.isnull().sum()
```

```python
df['bath'].fillna(df['bath'].median,inplace=True)
```

```python
df['balcony'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python

# total_sqft cleanup
def convert_sqft(x):
  if isinstance(x,str) and '-' in x:
    a,b = x.split('-')
    return (float(a)+float(b))/2
  try:
    return float(x)
  except:
    return None

df['total_sqft']=df['total_sqft'].apply(convert_sqft)
df.dropna(subset=['total_sqft'],inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```

Use the specific visualization types mentioned for each question.

Create a Heatmap to visualize correlations among numeric features (price, total_sqft, bath, balcony, etc.).


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(data=df,x=df['price'],kde=True,bins=30)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df,x="total_sqft",y="price")
plt.title('Relationship between Total Square Footage and Price')
plt.xlabel('Total Square Footage')
plt.ylabel('Price')
```

```python
sns.boxplot(data=df,x="bath",y="price")
plt.title('Effect of Bathrooms on Price')
plt.xlabel('Bathrooms')
plt.ylabel('Price')
plt.show()
```

```python
top_locations = df.groupby("location")["price"].mean().nlargest(10)

# Bar chart
sns.barplot(x=top_locations.values, y=top_locations.index)

plt.title("Top 10 Most Expensive Locations (Average Price)")
plt.xlabel("Average Price")
plt.ylabel("Location")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure()
plt.imshow(corr, interpolation='nearest')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap (Numeric Columns)")
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
# Select numeric columns
# numeric_cols = ['price', 'total_sqft', 'bath', 'balcony']
# Convert columns to numeric where possible
for col in ["price", "total_sqft", "bath", "balcony"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

numeric_cols = ["price", "total_sqft", "bath", "balcony"]

# Compute correlation matrix
corr_matrix = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
